```python
# Importing the libraries
import time
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.impute import SimpleImputer
from sklearn.metrics import precision_score, recall_score, f1_score,accuracy_score
# Importing the dataset
dataset = pd.read_csv('drive/MyDrive/creditcard.csv')

dataset.dropna(subset=['Class'], inplace=True)

y = dataset['Class']  # Replace 'TargetColumnName' with the actual name of your target column
X = dataset.drop(columns=['Class'])

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X.iloc[:, :] = imputer.fit_transform(X)


# Splitting the dataset into the Training set and Test set
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# K-Nearest Neighbors (K-NN)
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sn.heatmap(cm, annot=True, fmt="d", center=0)



# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# Naive Bayes
# Training the Naive Bayes model on the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sn.heatmap(cm, annot=True, fmt="d", center=0)

# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.svm import SVC
classifier_svm = SVC(kernel='rbf', C=1.0, gamma='scale')
classifier_svm.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier_svm.predict(X_test)


# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sn.heatmap(cm, annot=True, fmt="d", center=0)

# Visualising the Training set results
# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
